```python
# import the modules
import os
from os import listdir

names = []

# get the path/directory
folder_dir = "CAT_00"
for images in os.listdir(folder_dir):

    # check if the image ends with png
    if (images.endswith(".jpg")):
        names.append(images)
```

```python
import numpy as np

def add_gaussian_noise(image, mean=0, sigma=25):
    # Generate Gaussian noise
    row, col, ch = 100, 100, 3
    gauss = np.random.normal(mean, sigma, (row, col, ch))
    
    # Add the noise to the image
    noisy_image = image + gauss
    
    # Clip the values to be between 0 and 255 and convert back to uint8
    noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)
    
    return noisy_image, gauss

# Load your image
img = Image.open(f"CAT_00/{names[0]}")
img = img.resize((100, 100))

noisy_img = img
noise = np.zeros((100, 100, 3))

for i in range(20):
    noisy_img, gauss = add_gaussian_noise(noisy_img, sigma=50)
    noise += gauss

    imgplot = plt.imshow(noisy_img)
    plt.show()

print(noise / 20)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class GaussianDenoiser(nn.Module):
    def __init__(self):
        super(GaussianDenoiser, self).__init__()
        
        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1), # (3, 100, 100) -> (32, 100, 100)
            nn.ReLU(),
            nn.MaxPool2d(2, 2),                         # (32, 100, 100) -> (32, 50, 50)
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2)                          # (64, 50, 50) -> (64, 25, 25)
        )
        
        # Decoder
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, kernel_size=2, stride=2), # (64, 25, 25) -> (32, 50, 50)
            nn.ReLU(),
            nn.ConvTranspose2d(32, 3, kernel_size=2, stride=2),  # (32, 50, 50) -> (3, 100, 100)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# Instantiate the model
model = GaussianDenoiser()
print(model)
```

```python
import torch

def noisy_tensor(noisy_img=noisy_img):
    noisy_tensor = torch.from_numpy(noisy_img).float()
    noisy_tensor = noisy_tensor.permute(2, 0, 1)
    noisy_tensor = noisy_tensor.unsqueeze(0)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    noisy_tensor = noisy_tensor.to(device)

    return noisy_tensor

prediction = model(noisy_tensor())
```

```python
import random
import torch.optim as optim

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

for name in names:
    img = Image.open(f"CAT_00/{name}")
    img = img.resize((100, 100))

    noisy_img = img
    noise = np.zeros((100, 100, 3))

    random_number = random.randint(1, 20)

    for i in range(random_number):
        noisy_img, gauss = add_gaussian_noise(noisy_img, sigma=50)
        noise += gauss

    # imgplot = plt.imshow(noisy_img)
    # plt.show()

    scaled_noise = noise / random_number
    # print(scaled_noise)

    model.train()
    optimizer.zero_grad()

    predicted_noise = model(noisy_tensor(noisy_img))

    loss = criterion(predicted_noise.squeeze(0), torch.from_numpy(scaled_noise).float().permute(2, 0, 1).to(predicted_noise.device))
    loss.backward()
    optimizer.step()

    print(loss.item())
```
